```python
def count_files_in_folder(folder_path):
    
    '''
    Count the number of files in a folder
    '''

    file_count = 0
    for root, dirs, files in os.walk(folder_path):
        file_count += len(files)
    return file_count

folder_path = 'VGGFace2'  # Replace with the path to your folder
print(f'Total number of files: {count_files_in_folder(folder_path)}')
```

```python
def count_folders_in_folder(folder_path):

    '''
    Count the number of folders in a folder
    '''

    folder_count = 0
    for root, dirs, files in os.walk(folder_path):
        folder_count += len(dirs)
    return folder_count

folder_path = 'VGGFace2/val'  # Replace with the path to your folder
print(f'Total number of folders (val): {count_folders_in_folder(folder_path)}')

folder_path = 'VGGFace2/train'  # Replace with the path to your folder
print(f'Total number of folders (train): {count_folders_in_folder(folder_path)}')
```
